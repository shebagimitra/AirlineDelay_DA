# Number of flights in 2022

# Average flight delays in 2022

```python
avg_flight_delays = flights_of_2022.copy()
avg_flight_delays = avg_flight_delays.drop(avg_flight_delays.columns[[7, 8, 9, 10, 11, 12, 13, 14, 16, 17, 18, 19, 20]], axis=1)
avg_flight_delays
```

```python
avg_flight_delays['average_flight_delay'] = avg_flight_delays.apply(lambda row: row['arr_delay'] / row['arr_flights'] if row['arr_delay'] != 0 else float('NaN'), axis=1)
avg_flight_delays #This gives us the Number of minutes every flight has delayed on average.
```

# Average flight delays from 2022 to 2003

```python
avg_flight_delays_in_history = df.copy()
avg_flight_delays_in_history = avg_flight_delays_in_history.drop(avg_flight_delays_in_history.columns[[7, 8, 9, 10, 11, 12, 13, 14, 16, 17, 18, 19, 20]], axis=1)
avg_flight_delays_in_history
```

# List of all Unique Airlines

```python
start_year = 2022
end_year = 2012
year_range = df.loc[(df['year'] >= end_year) & (df['year'] <= start_year)]
year_range
```

```python
filter_value = 'Endeavor Air Inc.'
EAC_in_range = year_range.loc[year_range['carrier_name'] == filter_value]
max_endeavor_arr_flights = EAC_in_range['arr_flights'].max()
max_endeavor_arr_flights #A way to get the max arr_flight value for one airlines
```

```python
max_arr_flights = df.groupby("carrier_name")["arr_flights"].max() #Using groupby function to implement the same thing for all airlines
max_arr_flights
```

```python
max_arr_flights.max() #To see the maximum value in the list
```

```python
max_arr_flights[max_arr_flights == max_arr_flights.max()].index[0] #To see the associated airline with the maximum value
```
